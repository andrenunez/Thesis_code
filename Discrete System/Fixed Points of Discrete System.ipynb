## Trait 2 Dominates (Genetic Advantage)  ##

```python
x1 = 1
x2 = 2
y1 = 1
y2 = 2

def two_trait(x):
    return (x1*y1*(x**2)+ 0.5*(x1*y2+x2*y1)*x*(1-x))/((x1*x+x2*(1-x))*(y1*x+y2*(1-x)))

u1 = 0.5
u2 = 1-u1

curr = u1

arr = []

## Number of iterations in time ##
t = 100
x = [i for i in range(0,t+1)]
arr.append(u1)

for i in range(t):
    two_trait(curr)
    curr = two_trait(curr)
    arr.append(curr)
    
fig = plt.figure(figsize=(20,10))
plt.xlabel("Time (t)")
plt.ylabel("Population of trait 1 (% of Total Population)")
plt.title("Population of trait 1 over time, with parameters $a_{1} = 1,a_{2} = 2, b_{1} = 1,b_{2} = 2$")
plt.plot(x,arr)
plt.show()
fig.savefig("n2Trait2Adv.png")

```
