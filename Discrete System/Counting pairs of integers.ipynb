```python
## Creating the terms for the n-case model ##

import math

a = int(input("Number you want: "))

b = [i for i in range(1,a+1)]
c = [i for i in range(1,a+1)]

cdict = {}
pdict = {}

for k in range(1,a+1):
    cdict[k] = []
    pdict[k] = 0
    
counter = 0
counter_even = 0
counter_f= 0
counter_c = 0

for i in b:
    for j in c:
        if (i+j)/2 in c:
            counter +=1
            counter_even +=1
            #cdict[int((i+j)/2)].append((i,j))
            s1 = "n+"
            s2 = "n+"
            if int(i - (i+j)/2) < 0:
                s1 = "n"
            if int(j - (i+j)/2) < 0:
                s2 = "n"
            a = int(i - (i+j)/2)
            b = int(j - (i+j)/2)
    
            cdict[int((i+j)/2)].append(((s1 + str(a)),s2 + str(b)))
            pdict[int((i+j)/2)] += 1
            
            
        if math.floor((i+j)/2) in c and not ((i+j)/2).is_integer():
            counter +=1
            counter_f +=1
            #cdict[math.floor((i+j)/2)].append((i,j))
            
            s1 = "n+"
            s2 = "n+"
            
            if int(i-math.floor((i+j)/2)) < 0:
                s1 = "n"
            if int(j-math.floor((i+j)/2)) < 0:
                s2 = "n"
            
            a = int(i-math.floor((i+j)/2))
            b = int(j-math.floor((i+j)/2))
            
            
            cdict[math.floor((i+j)/2)].append(((s1 + str(a)),s2 + str(b)))
            pdict[math.floor((i+j)/2)] += 1
        if math.ceil((i+j)/2) in c and not ((i+j)/2).is_integer():
            
            s1 = "n+"
            s2 = "n+"
            
            if int(i-math.ceil((i+j)/2)) < 0:
                s1 = "n"
            if int(j-math.ceil((i+j)/2)) < 0:
                s2 = "n"
            
            a = int(i-math.ceil((i+j)/2))
            b = int(j-math.ceil((i+j)/2))
            
            cdict[math.ceil((i+j)/2)].append(((s1 + str(a)),s2 + str(b)))
            pdict[math.ceil((i+j)/2)] += 1
            counter +=1
            counter_c +=1
            
print("Total Terms: " + str(counter))
print("Even: " + str(counter_even))
print("Floor: " + str(counter_f))
print("Ceil: " + str(counter_c))
print(50*"*")
for i in cdict:
    print(str(i)+ ": " + str(cdict[i]))
print(pdict)
```
